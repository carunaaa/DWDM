```python
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
from IPython.display import display_html

def toy_dataset():
    value = np.array([[2, 5], [6, 3], [3, 4], [7, 8], [9, 11], [11, 14], [11, 17],
                  [55, 20], [25, 30], [50, 25], [49, 22], [54, 23], [53, 12], [43, 24], [52, 21],
                  [48, 50], [49, 28], [48, 60], [49, 50], [43, 51], [45, 55], [50, 48], [49, 60],
                  [40, 45], [45, 51], [47, 53], [48, 52], [30, 75], [110, 35], [160, 85]])
    titles = ['x', 'y']
    data = pd.DataFrame(value, columns=titles)
    
    print("Do you want to view data?")
    option = input()
    if option.lower() == 'yes':
        print("First five data points:")
        display_html(data.head())
    
    print("Do you want to view the scatter plot of data?")
    option = input()
    if option.lower() == 'yes':
        print("Data points scatter plot:")
        data.plot.scatter(x='x', y='y')
    
    return data

def Dbscan_clustering(data):
    db = DBSCAN(eps=10.5, min_samples=4).fit(data)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = pd.DataFrame(db.labels_, columns=['Cluster ID'])
    result = pd.concat((data, labels), axis=1)
    result.plot.scatter(x='x', y='y', c='Cluster ID', colormap='jet')

def main():
    data = toy_dataset()
    print("Clusters constructed by DBSCAN:")
    Dbscan_clustering(data)

main()
```
